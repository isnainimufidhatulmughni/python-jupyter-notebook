Partikel bermuatan __q__ bergerak dengan kecepatan

\begin{equation}
\vec{v}(t)=v_x(t)\hat{i} + v_y(t)\hat{j}
\end{equation}
$$\vec{v}(t)=v_x(t)\hat{i} + v_y(t)\hat{j}$$

dalam ruang bermedan magnetik konstan

$$\vec{B}=-\hat{k}B_z$$

Tentukan gerak partikel

## Question 1

## Answer 1

$$\sum F = ma$$


\begin{equation}
q\hat{v}\times \hat{B}= m \frac{d\hat{v}}{dt}
\tag{1}
\end{equation}

dimana 
\begin{equation}
\vec{v}(t)=v_x(t)\hat{i} + v_y(t)\hat{j}
\tag{2}
\end{equation}
dan $\vec{B}=-\hat{k}B_z$, sehingga:

$$q(v_x\hat{i} + v_y\hat{j})\times -\hat{k}B_z = m \frac{d}{dt} (v_x\hat{i} + v_y\hat{j})$$

\begin{equation}
q(v_x\hat{i} + v_y\hat{j})\times B_z = m(\frac{dv_x}{dt}\hat{i} + \frac{dv_y}{dt}\hat{j})
\tag{3}
\end{equation}

## Question 2

## Answer 2

Persamaan differensial dapat diperoleh dengan menyelesaikan perkalian cross pada ruas kiri:

Medan magnet $\vec{B}$ yang masuk ke bidang dapat dituliskan

\begin{equation}
\vec{B} = -B\hat{k}
\tag{4}
\end{equation}

\begin{equation}
\vec{v} \times \vec{B} = -\hat{i}Bv_y + \hat{j}Bv_x
\tag{5}
\end{equation}

Substitusi persamaan (5) ke persamaan (3):
$$q(\vec{v} \times \vec{B}) = q(Bv_x\hat{j}-Bv_y\hat{i}) = m(\frac{dv_x}{dt}\hat{i} + \frac{dv_y}{dt}\hat{j})$$

untuk arah-y:

$$qBv_x = m \frac{dv_y}{dt}$$

\begin{equation}
\frac{dv_y}{dt} - \frac{qB}{m}v_x = 0
\tag{6}
\end{equation}

Untuk arah-x:
$$-qBv_x = m \frac{dv_y}{dt}$$
\begin{equation}
\frac{dv_y}{dt} + \frac{qB}{m}v_x = 0
\tag{7}
\end{equation}

## Question 3

Selesaikan kedua persamaan diferensial sehingga dapat dieperoleh $V_x(t), V_y(t), x(t),$ dan  $y(t)$. Lakukan secara teori

## Answer 3

Substitusi persamaan (7) ke persamaan (6):

$$\frac{d}{dt}(-\frac{m}{qB}\frac{dv_x}{dt}) = \frac{qB}{m}v_x$$

$$-\frac{m}{qB}\frac{d^2v_x}{dt^2} = \frac{qB}{m}v_x$$

\begin{equation}
\frac{d^2v_x}{dt^2} + (\frac{qB}{m})^2v_x = 0
\tag{8}
\end{equation}

misal $k=\frac{qB}{m}$, maka persamaan (8) menjadi:
$$\frac{d^2v_x}{dt^2} + k^2 v_x = 0$$
yang merupakan persmaan osilasi, dengan solusi:

$$v_x=A \sin{kt} + B\cos{kt}$$

Substitusi persamaan (6) ke persamaan (7):

$$\frac{d}{dt}(\frac{m}{qB}\frac{dv_y}{dt}) = -\frac{qB}{m}v_y$$

$$\frac{m}{qB}\frac{d^2v_y}{dt^2} = -\frac{qB}{m}v_y$$

$$\frac{d^2v_y}{dt^2} + (\frac{qB}{m})^2v_y = 0$$

karena $k=\frac{qB}{m}$, maka
$$\frac{d^2v_y}{dt^2} + k^2 v_y = 0$$
yang merupakan persmaan osilasi, dengan solusi:

$$v_y=C \sin{kt} + D\cos{kt}$$

Saat $t=0$, misal partikel hanya bergerak horizontal, maka
$$t=0 -> v_y=0$$

$$0=C\sin{0} + D\cos{0}$$
$$D=0$$

Saat $t=\Delta{t}$, partikel bergerak vertikal saja:

$$v_x(\Delta{t})=0$$

$$0=A\sin{(k\Delta{t})} + B\cos{(k\Delta{t})}$$
$$\tan{(k\Delta{t})}=-\frac{B}{A}$$, Maka
$$v_x=A\sin{(k\Delta{t})} - Atan{(k\Delta{t})}\cos{(k\Delta{t})}$$

\begin{equation}
v_x = A(\sin{(k\Delta{t})} - tan{(k\Delta{t})}\cos{(k\Delta{t})})
\tag{9}
\end{equation}

dan
\begin{equation}
v_y = C\sin{(k\Delta{t})}
\tag{10}
\end{equation}

Sementara saat $t=0, v_x$ harus bernilai misalkan u

$$v_x(0) = u = A(\sin{0} - \tan{(k\Delta {t})} \cos{0})$$

$$u = A \tan{(k\Delta {t})}$$ Maka

\begin{equation}
v_x(t) = \frac{u}{\tan{(k\Delta {t})}} \sin{(k\Delta{t})} - u \cos{(k\Delta{t})}
\tag{11}
\end{equation}


Persamaan posisi dapat diperoleh dengan integral persamaan (10) dan persamaan (11) terhadap t

\begin{equation}
y(t) = \int C\sin{(k\Delta{t})}\,dt
\tag{12}
\end{equation}

$$x(t) = \int \frac{u}{\tan{(k\Delta {t})}} \sin{(k\Delta{t})} - u \cos{(k\Delta{t})}\,dt$$

$$x(t) = -\frac{u}{k\tan{(k\Delta {t})}} \cos{(k\Delta{t})} - \frac{u}{k} \sin{(k\Delta{t})} + x_0$$

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 30, 0.05)
vx = -15*np.sin(t)
vy = 15*np.cos(t)

plt.figure()
plt.plot(t, vx, label='$v_x$')
plt.plot(t, vy, label='$v_y$')
plt.title('Rotasi Partikel pada Medan Magnet (Teoritik)')
plt.xlabel('Waktu')
plt.ylabel('Kecepatan')
plt.legend(loc=1)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Persamaan Runge-Kutta Orde 4
def RK4(F, t, x, y, h):
    K0 = h * F(t, x, y)
    K1 = h * F(t + h/2.0, x + K0/2.0, y)
    K2 = h * F(t + h/2.0, x + K1/2.0, y)
    K3 = h * F(t + h, x + K2, y)
    return(K0 + 2.0 * K1 + 2.0 * K2 + K3)/6.0

# Persamaan Differensial Terkopel
def vx_kop(t, vx, vy, q = 1, Bz = 1, m = 1):
    return -q * Bz/m * vy

def vy_kop(t, vy, vx, q = 1, Bz = 1, m = 1):
    return q * Bz/m * vx

# Input Parameter
vx0 = 0
vy0 = 15

t0 = 0
t_stop = 60

h = 0.005

vxs=[vx0]
vys=[vy0]
ts=[t0]

# Grafik solusi dengan metode RK4
while ts[-1] < t_stop:
    h = min(h, t_stop - ts[-1])
    vx = vxs[-1] + RK4(vx_kop,ts[-1],vxs[-1],vys[-1],h)
    vy = vys[-1] + RK4(vy_kop,ts[-1],vys[-1],vxs[-1],h)
    t = ts[-1] + h
    
    ts.append(t)
    vxs.append(vx)
    vys.append(vy)

plt.figure()
plt.grid()
plt.plot(ts, vys, 'r', label = '$v_x$')
plt.plot(ts, vxs, 'b', label = '$v_y$')
plt.title('Rotasi partikel pada medan magnet dengan metode Runge-Kutta')
plt.xlabel('Waktu (s)')
plt.ylabel('Kecepatan')
plt.legend(loc = 'upper left')
plt.savefig('Pakai RK4')
plt.show()
```
